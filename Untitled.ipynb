```python
import pandas as pd
from random import randint 
from sklearn import preprocessing
from sklearn.preprocessing import MinMaxScaler
import keras
import matplotlib.pyplot as plt
from keras import backend as K
from keras.models import Sequential
from keras.layers import Activation
from keras.layers.core import Dense
from keras.optimizers import Adam,SGD
from keras.metrics import categorical_crossentropy
from keras.models import model_from_json
import os
df = pd.read_csv('data5.csv')
X = df.iloc[0:39,1:11]
columns = df.columns
Xsample = X
train_label = df.iloc[0:39,0]

train_sample = Xsample
scaler = MinMaxScaler(feature_range=(0,1))
le = preprocessing.LabelEncoder()
train_sample = train_sample.apply(le.fit_transform)
scaled_train_samples = scaler.fit_transform((train_sample))
model = Sequential([
    Dense(10,input_shape=(10,),activation='softmax'),
    Dense(30,activation='softmax'),
    Dense(30,activation='softmax'),
    Dense(30,activation='softmax'),
    Dense(2,activation='softmax')
])
model.compile(Adam(lr=.0001), loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
history=model.fit(scaled_train_samples,train_label,batch_size=5,epochs=3000, shuffle=True, verbose=2,validation_split=0.2)
```

```python
X = df.iloc[39:44,1:11]
le = preprocessing.LabelEncoder()
X = X.apply(le.fit_transform)
train_label = df.iloc[39:44,0]
scores = model.evaluate(X,train_label, verbose=0)
print("%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```

```python
X = df.iloc[0:44,1:14]
le = preprocessing.LabelEncoder()
X = X.apply(le.fit_transform)
predic = model.predict(X,batch_size=4,verbose=0)
for i in predic:
    print(i)
```
