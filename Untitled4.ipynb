```python
import pandas as pd
from random import randint 
from sklearn import preprocessing
from sklearn.preprocessing import MinMaxScaler
import keras
from keras.utils import to_categorical
import matplotlib.pyplot as plt
from keras import backend as K
from keras.models import Sequential
from keras.layers import Activation
from keras.layers.core import Dense
from keras.optimizers import Adam,SGD
from keras.metrics import categorical_crossentropy,binary_crossentropy
from keras.models import model_from_json
import os
df = pd.read_csv('data3.csv')
X = df.iloc[0:39,1:9]
columns = df.columns
Xsample = X
train_label = df.iloc[0:39,0]
```

```python
train_sample = Xsample
scaler = MinMaxScaler(feature_range=(0,1))
le = preprocessing.LabelEncoder()
train_sample = train_sample.apply(le.fit_transform)
scaled_train_samples = scaler.fit_transform((train_sample))
model = Sequential([
    Dense(8,input_shape=(8,),activation='relu'),
    Dense(24,activation='relu'),
    Dense(24,activation='relu'),
    Dense(24,activation='relu'),
    Dense(2,activation='softmax')
])
```

```python
model.summary()
```

```python
model.compile(Adam(lr=.0001), loss='sparse_categorical_crossentropy',metrics=['accuracy'])
history=model.fit(scaled_train_samples,train_label,batch_size=4,epochs=10000, shuffle=True, verbose=2,validation_split=0.2)
```

```python
X = df.iloc[39:44,1:9]
le = preprocessing.LabelEncoder()
X = X.apply(le.fit_transform)
train_label = df.iloc[39:44,0]
scores = model.evaluate(X,train_label, verbose=0)
print("%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```
